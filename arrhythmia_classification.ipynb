```python
import tensorflow.keras.backend as K

def recall_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

def precision_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
def get_ecg_list_cic():
    with open(os.path.join(dirpath, 'REFERENCE.csv'), newline='') as csvfile:
        csvreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        r = list()
        for row in csvreader:
            d = loadmat(os.path.join(dirpath, row[0]+'.mat'))
            if row[1] == 'N':
                y = 0
            elif row[1] == 'A':
                y = 1
            elif row[1] == 'O':
                y = 2
            elif row[1] == '~':
                y = 3
            else:
                assert False, 'Unknown class %s.' % row[0]
            r.append([row[0], y, d['val'][0]])
    return(r)
import itertools as it

def split_dataset_cic_original(ecg_list, training_ratio=0.8): # dataset을 training set과 test set으로 나눈다
    training_x = list()
    training_y = list()
    test_x = list()
    test_y = list()
    for row in ecg_list:
        if len(row[2]) == 9000:# 전체 8528개의 데이터중 5977개가 30초짜리 샘플이다. 이 데이터만 쓴다. (9000개의 샘플)
            x = list()
            for i in range(15):
                timedistributed_x = list()
                for j in range(600):
                    timedistributed_x.append([row[2][i*600+j]])
                x.append(timedistributed_x)
            if random.random() > training_ratio: #TO DO: the same patients should belong to the same train/test sets
                test_x.append(x)
                test_y.append(row[1])
            else:
                training_x.append(x)
                training_y.append(row[1])
    training_x = np.array(training_x)
    training_y = np.array(training_y)
    test_x = np.array(test_x)
    test_y = np.array(test_y)
    training_y = to_categorical(training_y, num_classes=4)
    test_y = to_categorical(test_y, num_classes=4)
    return (training_x, training_y, test_x, test_y)

def ReshapeMat (mat):
    sh= np.array(mat.shape)
    return tf.reshape(mat,[sh[0], sh[2], sh[3] ,sh[1]] )# e.g. reshape [2928, 1, 22, 512] to [2928, 22, 512, 1]

def maketensor(x):
    x2=tf.convert_to_tensor(x,dtype=tf.float64)
    return(x2)

def split_dataset_cic(ecg_list, testset_size=0.3, window_size=512, step_size=256): 
    row0 = list()
    row1 = list()
    row2 = list()
    for row in ecg_list:
        if len(row[2]) == 9000:# 전체 8528개의 데이터중 5977개가 30초짜리 샘플이다. 이 데이터만 쓴다. (9000개의 샘플)
            signal= row[2]
            resampled  = resample(signal, 6000) # downsample 300Hz(9000) to 200Hz(6000)
            # extract windows
            initindx=np.array(range(0, 6000-window_size, step_size))
            windows_perPatient = list()
            for i in initindx: 
                windows_perPatient.append(resampled[i:i+window_size])
            row1.append(row[1]) # label
            row0.append(row[0]) # ID 
            row2.append([windows_perPatient])
    row2 = ReshapeMat(np.array(row2))           
    
    # stratified sampling
    ID_train0,ID_test,y_train0,y_test,X_train0,X_test = train_test_split(row0,np.array(row1),np.array(row2), test_size=testset_size, stratify=row1,random_state=42)
    ID_train,ID_val,y_train,y_val,X_train,X_val = train_test_split(ID_train0,y_train0,X_train0, test_size=testset_size, stratify=y_train0,random_state=42)
    
    # one-hot encoding
    y_train_catg = to_categorical(y_train, num_classes=4)
    y_val_catg = to_categorical(y_val, num_classes=4)
    y_test_catg = to_categorical(y_test, num_classes=4)
    
    return (maketensor(X_train), maketensor(X_val), maketensor(X_test), 
            maketensor(y_train_catg), maketensor(y_val_catg), maketensor(y_test_catg))

# https://buomsoo-kim.github.io/keras/2018/05/02/Easy-deep-learning-with-Keras-8.md/
# https://keras.io/layers/convolutional/

def get_cic_model(METRICS,HP2,HP3=False,HP4=False):

    model = Sequential()
    if HP3:
        model.add(BatchNormalization())
        model.add(Dropout(0.5))
        
    model.add(TimeDistributed(Conv1D(filters=8, kernel_size=5, activation='relu', padding='valid', strides=1),input_shape = (22,512,1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=2, strides=2)))
    if HP3:
        model.add(BatchNormalization())
        model.add(Dropout(0.5))
        
    model.add(TimeDistributed(Conv1D(filters=16, kernel_size=5, activation='relu', padding='valid', strides=1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=2, strides=2)))
    if HP3:
        model.add(BatchNormalization())
        model.add(Dropout(0.5))
        
    model.add(TimeDistributed(Conv1D(filters=32, kernel_size=5, activation='relu', padding='valid', strides=1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=2, strides=2)))
    if HP3:
        model.add(BatchNormalization())
        model.add(Dropout(0.5))
        
    model.add(TimeDistributed(Conv1D(filters=64, kernel_size=5, activation='relu', padding='valid', strides=1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=2, strides=2)))
    if HP3:
        model.add(BatchNormalization())
        model.add(Dropout(0.5))
        
    model.add(TimeDistributed(Conv1D(filters=128, kernel_size=5, activation='relu', padding='valid', strides=1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=2, strides=2)))
    if HP3:
        model.add(BatchNormalization())
        model.add(Dropout(0.5))
        
    model.add(TimeDistributed(Conv1D(filters=256, kernel_size=5, activation='relu', padding='valid', strides=1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=2, strides=2)))
    if HP3:
        model.add(BatchNormalization())
        model.add(Dropout(0.5))
        
    model.add(TimeDistributed(GlobalAveragePooling1D()))
    if HP3:
        model.add(BatchNormalization())
        model.add(Dropout(0.5))
        
    model.add(LSTM(128)) 
    if HP4:
        model.add(LSTM(128)) 
    model.add(Dense(4, activation='softmax'))
#     https://www.tensorflow.org/api_docs/python/tf/keras/Sequential?version=stable
    # When using the delayed-build pattern (no input shape specified), you can
    # choose to manually build your model by calling `build(batch_input_shape)`:
    model.build((None, 22, 512, 1)) 
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3), loss=tf.keras.losses.BinaryCrossentropy(label_smoothing=HP2), 
                  metrics=METRICS) 
    return(model)
    

def scoring(test_y,pred_y):
    res = np.zeros((4, 4))

    for i in range(len(pred_y)):
        res[test_y[i], pred_y[i]] += 1

    print(res)
    score = list()

    for i in range(4):
        precision = res[i, i] / sum(res[:, i])
        recall = res[i, i] / sum(res[i, :])
        f1 = harmonic_mean([precision, recall])
        score.append([precision, recall, f1])

    return(np.array(score))
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        cm = cm * 100
        print("\nNormalized confusion matrix")
    else:
        print('\nConfusion matrix, without normalization')
        print(cm)
        print ()

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.0f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```
